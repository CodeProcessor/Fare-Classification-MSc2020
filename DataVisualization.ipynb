```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_filename = 'data/train.csv'
train_df = pd.read_csv(train_filename)
train_df.dropna(inplace=True)
train_df.info()
```

```python

sns.lmplot("duration", "fare", data=train_df, fit_reg=False, col='label', col_wrap=2, ci=5)
```

```python
useful_colomns_2 = ["additional_fare", "duration", "meter_waiting", "meter_waiting_fare", "fare"]
df1 = train_df[useful_colomns_2]
# df1['Produced Qty'] = df1['Produced Qty'].astype(float)
corr = df1.loc[:,df1.dtypes == 'float64'].corr()
ax = plt.axes()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, 
            cmap=sns.diverging_palette(220, 10, as_cmap=True), annot=True, ax = ax)

```
